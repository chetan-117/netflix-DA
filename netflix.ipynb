```python
# find 'House of Cards', then its show id and its directors
temp_dataset = dataset.set_index('Title')

result_row = temp_dataset.loc['House of Cards']

show_id = result_row['Show_Id']
director = result_row['Director']

print(show_id)
print(director)
```

### in which year most number of TV shows and Movies were released

```python
# creating a bar graph of this
dataset['Dates'].dt.year.value_counts().plot(kind='bar')
```
